```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.azlyrics.com/f/fleetwood.html'

lyrics_page = requests.get(url)

soup = BeautifulSoup(lyrics_page.content, 'lxml')
lyrics_links = soup.find('div', id="listAlbum").find_all('a', href=True)

for link in lyrics_links:
    url = f'https://www.azlyrics.com{link["href"]}'

    lyrics_page = requests.get(url)

    soup = BeautifulSoup(lyrics_page.content, 'lxml')
    lyrics_div = soup.find('div', class_="ringtone").find_next('div')

    print(lyrics_div.text)
```
